```python
from pyndamics import Simulation
from pyndamics.emcee import *
```

```python
t=linspace(0,10,100)
h=3.4*exp(.3*t)

sim=Simulation()
sim.add("h'=a*h",1,plot=True)
sim.add_data(t=t,h=h,plot=True)
sim.params(a=1)
sim.run(0,10)
```

```python
model=MCMCModel(sim,
                a=Normal(0,10),
                initial_h=Uniform(0,100),
                )
```

```python
model.run_mcmc(500)
model.set_initial_values('samples')
```

```python
model.plot_distributions()
```
