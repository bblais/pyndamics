```python
from pyndamics import Simulation
```

```python
sim=Simulation()
sim.add("y'=y - 1/2*exp(t/2)*sin(5*t) + 5*exp(t/2)*cos(5*t)",0)
sim.run(0,5)
```

```python
t,y=sim.t,sim.y
plot(t,y)
y2=exp(t/2)*sin(5*t)
plot(t,y2,'r--')
```
